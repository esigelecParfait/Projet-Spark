```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("CyclistesLogs").getOrCreate()

df = spark.read.option("header", True).csv("Cycliste/")
```

```python
df = df.dropDuplicates(["timestamp"])
```

```python
df.count()
```

```python
df.show(5)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import col, lag

window = Window.partitionBy("id").orderBy("timestamp")
```

```python
df = df.withColumn(
    "prev_sur_velo",
    lag("sur_velo").over(window)
)
```

```python
from pyspark.sql.functions import when

df = df.withColumn(
    "start_trip",
    when(
        (col("sur_velo") == True) & (col("prev_sur_velo") == False),
        1
    ).otherwise(0)
)
```

```python
from pyspark.sql.functions import sum as spark_sum

df = df.withColumn(
    "trip_id",
    spark_sum("start_trip").over(window)
)
```

```python
from pyspark.sql.functions import to_timestamp

df = df.withColumn(
    "timestamp",
    to_timestamp("timestamp")
)
```

```python
from pyspark.sql.functions import min, max

trajets = df.filter(col("sur_velo") == True) \
    .groupBy("id", "trip_id") \
    .agg(
        min("timestamp").alias("start_time"),
        max("timestamp").alias("end_time")
    )
```

```python
trajets = trajets.withColumn(
    "duration",
    col("end_time") - col("start_time")
)
```

```python
from pyspark.sql.functions import unix_timestamp

df = df.withColumn(
    "timestamp",
    unix_timestamp("timestamp")
)
```

```python
cycliste_id = 246

cycliste_df = trajets.filter(trajets.id == cycliste_id)
```

```python
from pyspark.sql.functions import unix_timestamp, col

# Convert timestamp columns to Unix timestamps (long) and duration to seconds (long)
# before converting to Pandas DataFrame.
processed_cycliste_df = cycliste_df \
    .withColumn("start_time", unix_timestamp(col("start_time"))) \
    .withColumn("end_time", unix_timestamp(col("end_time"))) \
    .withColumn("duration", unix_timestamp(col("end_time")) - unix_timestamp(col("start_time")))

pdf = processed_cycliste_df.toPandas()
```

```python
pdf = pdf.sort_values("trip_id")
```

```python
import matplotlib.pyplot as plt

plt.figure()

plt.plot(pdf["trip_id"], pdf["duration"])

plt.xlabel("Numéro de trajet")
plt.ylabel("Durée")
plt.title(f"Durée des trajets du cycliste {cycliste_id}")

plt.show()
```
